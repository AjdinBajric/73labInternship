```python
import pandas as pd
import numpy as np
```

```python
def read_json(file):
    reader = pd.read_json(f"Data/{file}.json")
    return reader[['onetsoc_code','element_id', 'scale_id', 'data_value']]
```

```python
knowledge = read_json("knowledge")
abilities = read_json("abilities")
skills = read_json("skills")
work_activities = read_json("work_activities")
```

```python
df_abilities = abilities["onetsoc_code"].value_counts().sort_values()
df_knowledge = knowledge["onetsoc_code"].value_counts().sort_values()
df_skills = skills["onetsoc_code"].value_counts().sort_values()
df_work_activities = work_activities["onetsoc_code"].value_counts().sort_values()
```

```python
final_dataset = pd.DataFrame({
    "job_id": [],
    "missing_abilities": [],
    "missing_knowledge": [],
    "missing_skills": [],
    "missing_activities": [],
    "number_of_missing_values": []
})
```

```python
def add_to_dataframe(job_id, n, num_of_missing_val):
    global final_dataset
    final_dataset = final_dataset.append({"job_id":job_id, "missing_abilities": n==1, "missing_knowledge": n==2, "missing_skills": n==3, "missing_activities": n==4, "number_of_missing_values": num_of_missing_val}, ignore_index = True)
```

```python
missing_abilities = [ add_to_dataframe(df_abilities.index[i], 1, 104 - df_abilities[i]) for i in range(len(df_abilities)) if df_abilities[i] < 104]
missing_knowledge = [add_to_dataframe(df_knowledge.index[i], 2, 66 - df_knowledge[i]) for i in range(len(df_knowledge)) if df_knowledge[i] < 66]
missing_skills = [add_to_dataframe(df_skills.index[i], 3, 70 - df_skills[i]) for i in range(len(df_skills)) if df_skills[i] < 70]
missing_activities = [add_to_dataframe(df_work_activities.index[i], 4, 82 - df_work_activities[i]) for i in range(len(df_work_activities)) if df_work_activities[i] < 82]
```
